```python
import pandas as pd
import numpy as np
import json
```

```python
with open('adv_examples/many-to-one/adv_gen_l_inf.json', 'r') as f:
    data = json.loads(f.read())
```

```python
for dataset in data.keys():
    for model_name in data[dataset].keys():
        lines = []
        for adv_attack in data[dataset][model_name].keys():
            for setting in data[dataset][model_name][adv_attack].keys():
                line = {
                    "attack_type" : adv_attack,
                    "setting" : setting
                }
                line = {**line, **data[dataset][model_name][adv_attack][setting]["metrics"]}
                lines.append(line)
        print(model_name)
        display(pd.DataFrame(lines).sort_values("R2", ascending=False).set_index(["attack_type", "setting"]))
    
    
    break
```

```python
with open('adv_examples/many-to-one/adv_gen_l_inf_2.json', 'r') as f:
    data = json.loads(f.read())
    
with open('models/many-to-one/model_info.json', 'r') as f:
    originals = json.loads(f.read())

    
metrics = [
    "R2",
    "RMSE",
    "MAE",
    "MAPE",
]
```

```python
rows = []
for dataset in data.keys():
    for model_name in data[dataset].keys():
        for metric in metrics:
            rows.append((dataset,model_name,metric))
            

columns = [
    ('Original',''), \
    ('FGSM', '0.1'), ('FGSM', '0.05'), ('FGSM', '0.025'), ('FGSM', '0.01'), \
    ('PGD', '0.025 | 0.01'), ('PGD', '0.025 | 0.025'), ('PGD', '0.01 | 0.01'), ('PGD', '0.01 | 0.025')
]

row_list = pd.MultiIndex.from_tuples(rows)
col_list = pd.MultiIndex.from_tuples(columns)

A = pd.DataFrame(index = row_list, columns = col_list)
A
```

```python
for dataset in originals.keys():
    for model_name in originals[dataset].keys():
        for metric in metrics:
                    A.loc[(dataset, model_name, metric)][('Original', '')] = round(originals[dataset][model_name][metric] ,2)           
```

```python
for dataset in data.keys():
    for model_name in data[dataset].keys():
        for adv_attack in data[dataset][model_name].keys():
            for setting in data[dataset][model_name][adv_attack].keys():
                parsed_setting = setting.replace('epsilon=', '').replace('alpha=', '')
                for metric in metrics:
                    A.loc[(dataset, model_name, metric)][(adv_attack, parsed_setting)] = data[dataset][model_name][adv_attack][setting]["metrics"][metric]
```
